```python
"""

1. What is data cleaning in the context of data analytics?

Answer:
Data cleaning is the process of identifying and correcting errors, inconsistencies, duplicates, and missing values in a dataset to improve its quality before analysis. Clean data ensures accurate results and reliable decision-making.

Example: Removing duplicate customer records and filling missing age values before analyzing customer behavior.

2. What are some common techniques used for data cleaning?

Answer:
Common data cleaning techniques include:

Removing duplicate records
Handling missing values (mean, median, mode, or deletion)
Correcting inconsistent data formats
Detecting and removing outliers
Standardizing text (e.g., "Male", "male", "M")
Fixing incorrect or invalid values
Removing irrelevant columns
Data normalization and scaling

3. How does missing data affect data modeling?

Answer:
Missing data reduces the accuracy and reliability of machine learning models. It can lead to biased predictions, incorrect statistical analysis, and poor model performance.

Effects:

Lower prediction accuracy
Biased results
Reduced dataset size if rows are removed
Difficulty in training models

Solution: Use imputation methods such as mean, median, mode, or predictive imputation.

4. Explain the concept of outlier detection during data cleaning.

Answer:
Outlier detection is the process of identifying data points that are significantly different from the rest of the dataset. Outliers may result from errors or represent unusual observations.

Methods for detecting outliers:

Box Plot (IQR Method)
Z-Score
Scatter Plot
Isolation Forest (Machine Learning)

Example: If most employee salaries are ₹30,000–₹70,000 and one record shows ₹50,00,000, it may be an outlier.

5. What is the purpose of normalization in data preprocessing?

Answer:
Normalization scales numerical values to a common range (usually 0 to 1) so that features with larger values do not dominate smaller ones.

Benefits:

Improves machine learning model performance
Speeds up model training
Ensures equal importance of all features

Example:
Before normalization:

Age = 45
Salary = 8,00,000

After normalization:

Age = 0.45
Salary = 0.80

6. What are some strategies to handle categorical variables during data cleaning?

Answer:
Common strategies include:

Label Encoding
One-Hot Encoding
Ordinal Encoding
Frequency Encoding
Target Encoding (advanced)

Example:

Gender	Label Encoding
Male	1
Female	0

One-Hot Encoding:

City	Delhi	Mumbai	Pune
Delhi	1	0	0
<p>

7. Discuss the importance of feature engineering in data analytics.

Answer:
Feature engineering is the process of creating new features or modifying existing ones to improve model performance.

Importance:

Increases prediction accuracy
Helps models learn better patterns
Reduces complexity
Improves business insights

Example:
Instead of using Date of Birth, create a new feature:

Age = Current Year − Birth Year

This provides more useful information for prediction.

8. What is the curse of dimensionality and how does it affect data modeling?

Answer:
The curse of dimensionality refers to problems that arise when a dataset has too many features (dimensions).

Effects:

Increased computational cost
Higher memory usage
Model overfitting
Lower prediction accuracy
Sparse data distribution

Solution:

Feature Selection
Principal Component Analysis (PCA)
Remove irrelevant features

9. Provide an example of a data preprocessing technique used to handle imbalanced datasets.

Answer:
One common technique is SMOTE (Synthetic Minority Over-sampling Technique).

SMOTE creates synthetic samples for the minority class to balance the dataset.

Other techniques:

Random Oversampling
Random Undersampling
ADASYN
Class Weighting

Example:
Before balancing:

Fraud = 100
Non-Fraud = 10,000

After applying SMOTE:

Fraud = 10,000
Non-Fraud = 10,000

This improves the model's ability to detect the minority class.


10. What role does data quality play in the success of a data analytics project?

Answer:
Data quality is critical because analytics results are only as reliable as the data used. High-quality data leads to accurate insights, better predictions, and informed business decisions.

Characteristics of good data quality:

Accuracy
Completeness
Consistency
Validity
Timeliness
Uniqueness

Importance:

Improves model performance
Increases trust in analysis
Reduces errors
Supports better decision-making
Saves time and cost during analysis
"""
```
